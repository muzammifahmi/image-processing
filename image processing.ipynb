```python
# Load an image file from the local dataset folder
from PIL import Image
import matplotlib.pyplot as plt

# Example: Load and display 'Lena.png' from the dataset folder
image_path = 'dataset/Lena.png'
image = Image.open(image_path)
plt.imshow(image)
plt.axis('off')
plt.title('Lena.png')
plt.show()
```

```python
# Read PNG/JPG file from local dataset folder
import cv2
import matplotlib.pyplot as plt

# Set the filename to the image you want to read from the dataset folder
filename = 'dataset/trial1.jpg'  # Change to your desired file

# Read the image with OpenCV
img = cv2.imread(filename)

# Check if the image was loaded successfully
if img is not None:
    # Convert color (BGR -> RGB)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # Display
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title(filename)
    plt.show()
else:
    print(f"Failed to load image: {filename}")
```

```python
import pydicom
import matplotlib.pyplot as plt
import os

# Set the correct DICOM file path from the dataset folder
filename = 'dataset/1.dcm'

# Check if the file exists
if os.path.exists(filename):
    # Read the DICOM file
    ds = pydicom.dcmread(filename)
    # Get the pixel array
    img_dicom = ds.pixel_array
    # Display the image
    plt.imshow(img_dicom, cmap="gray")
    plt.axis("off")
    plt.title(filename)
    plt.show()
    # Show DICOM metadata
    print(ds)
else:
    print(f"File not found: {filename}")
```
