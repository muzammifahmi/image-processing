# Aritmatika

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img1 = cv2.imread('citra1.png')
img2 = cv2.imread('citra2.png')
```

teks yang dimiringkan# Bagian Baru

## Operasi Aritmatika

```python
#operasi penjumlahan
add_result = cv2.add(img1,img2)
```

```python
#operasi pengurangan
sub_result = cv2.subtract(img1,img2)
diff_result = cv2.absdiff(img1,img2)
```

```python
#operasi perkalian
mult_result = cv2.multiply(img1, img2)
```

```python
#operasi pembagian
div_result = cv2.divide(img1, img2)
```

```python
#operasi bitwise
and_result = cv2.bitwise_and(img1, img2)
or_result = cv2.bitwise_or(img1, img2)
xor_result = cv2.bitwise_xor(img1, img2)
not_result = cv2.bitwise_not(img1)
```

## Menampilkan Hasil

```python
#menampilkan hasil operasi
print('original')
cv2_imshow(img1)
cv2_imshow(img2)
print('penjumlahan')
cv2_imshow(add_result)
print('pengurangan')
cv2_imshow(sub_result)
cv2_imshow(diff_result)
print('perkalian')
cv2_imshow(mult_result)
print('pembagian')
cv2_imshow(div_result)
print('bitwise')
cv2_imshow(and_result)
cv2_imshow(or_result)
cv2_imshow(xor_result)
cv2_imshow(not_result)
```
