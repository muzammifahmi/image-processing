```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('noise.jpg')
cv2_imshow(img)
```

```python
mean = cv2.blur(img, (3,3))
gaussian = cv2.GaussianBlur(img, (5,5),0)
median = cv2.medianBlur(img, 3)

title = ['Citra Noise', 'Mean Filter', 'Gaussian Filter', 'Median Filter']
images = [img, mean, gaussian, median]

plt.figure(figsize = (8,8))
for i in range(4):
  plt.subplot(2,2,i+1)
  plt.imshow(images[i])
  plt.title(title[i])
  plt.axis('off')
plt.show()

```

```python
sobelx = cv2.Sobel(median, cv2.CV_64F, 1, 0, ksize = 3)
sobely = cv2.Sobel(median, cv2.CV_64F, 0, 1, ksize = 3)
prewittx = cv2.filter2D(median, -1, np.array([[-1,0,1],[-1,0,1],[-1,0,1]]))
prewitty = cv2.filter2D(median, -1, np.array([[1,1,1],[0,0,0],[-1,-1,-1]]))




print('median')
cv2_imshow(median)
print('sobelx')
cv2_imshow(sobelx)
print('sobely')
cv2_imshow(sobely)
print('prewittx')
cv2_imshow(prewittx)
print('prewitty')
cv2_imshow(prewitty)
```

```python
smoothing = cv2.bilateralFilter(median, 5, 75, 75)
sharpening = cv2.addWeighted(smoothing, 1.5, cv2.GaussianBlur(smoothing, (0,0), 10), -0.5, 0)
edge_detection = cv2.Canny(sharpening, 100, 200)


cv2_imshow(smoothing)
cv2_imshow(sharpening)
cv2_imshow(edge_detection)
```
