```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# membaca citragrayscale
img = cv2.imread('sample_data/fahmi.jpg', 0)

# melakukan histogram equalization
equalized_img = cv2.equalizeHist(img)

```

```python
# menampilkan hasil perbandingan
plt.figure(figsize=(12, 8))

#citra asli
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli')

# histogram citra asli
plt.subplot(2, 2, 2)
plt.hist(img.ravel(), 256, [0, 256])
plt.title('Histogram Asli')

# citra setelah equalization
plt.subplot(2, 2, 3)
plt.imshow(equalized_img, cmap='gray')
plt.title('Setelah Equalization')

# histogram setelah equalization
plt.subplot(2, 2, 4)
plt.hist(equalized_img.ravel(), 256, [0, 256])
plt.title('Histogram equalized')

plt.tight_layout()
plt.show()
```

```python
# menyimpan hasil
cv2.imwrite('sample_data/fahmi_equalized.jpg', equalized_img)
```
